<b>Adım Adım Yapılacaklar:</b><br>

<b>5)</b> <span style="color:blue"><b>ucgen2_secim</b></span> fonksiyonunun içindeki <span style="color:green"><b>else:</b></span> bloğunun içinde <span style="color:blue"><b>ucgen2_cizim</b></span> fonksiyonunu çağırınız. <b>(Parametre olarak inp değişkenini fonksiyonun içine yollamayı unutmayın!)</b><br>
<b>6)</b> <span style="color:blue"><b>main</b></span> adında bir fonksiyon tanımlayıp. parametre olarak üçgen tipini alıp<br>

&emsp;<b>b)</b> Eğer <span style="color:blue"><b>1</b></span> ise daha önceki örnekteki üçgeni seçmiş olacak. <b>(ucgen1_secim fonksiyonunu çağırın.)</b><br> 
&emsp;<b>c)</b> Eğer <span style="color:blue"><b>2</b></span> ise bu örnek için gösterdiğim yeni üçgenimizi seçecek. (ucgen2_secim fonksiyonunu çağırın.)<br>
&emsp;<b>d)</b> Diğer değerlerde ise <span style="color:darkred"><b>"HATA"</b></span> yazdıracak. <br>
<b>7)</b> Ardından kodumuzda ( tüm fonksiyonların dışında ) kullanıcıdan bir değer alıp bu değeri <span style="color:blue"><b>main</b></span> fonksiyonunu çağırıp o fonksiyona gönderin. <b>(main(inp))</b><br><br>
<b>8)</b> Çalıştırın ve programınızın doğru çalışıp çalıpmadığını test edin. Çalışıyorsa devam edin.<br>
<b>9)</b> <span style="color:blue"><b>main</b></span> fonksiyonunun çağrılması ve kullanıcıdan input alınması kısmını silelim.<br>

<br>
<b>11)</b> Yeni dosyanın oluşturulup oluşturulmadığını kontrol edin! Eğer oluşturulmuşsa devam edin.<br>
<b>11)</b> Tüm fonksiyonlarınız dosyaya kayıt edildiğine göre tüm kodunuzu silip <span style="color:cadetblue"><b>mylib</b></span> modülünden <span style="color:blue"><b>main</b></span> fonksiyonunu kodunuza ekleyerek programınızı çalıştırın.<br>










**Örnek - 1 'deki çözüm aşağıda verilmiştir.**<br>
**Lütfen kodunuzu 2. bloğa yazın.**

```python
while 1:
    inp = int(input("Satır sayısını giriniz: ")) #  Kullanıcıdan integer bir input aldık.
    
    if inp == 0:
        print("Quit")
        break
    elif inp == 1:
        print("Geçersiz INPUT")
    elif (inp < 0 or inp > 20):
        print("OVERFLOW")
    else:
        
        numbers = 1 # Üçgeni 1'den başlaktık.

        for row in range(inp): # Input kadar satır yazdırmak için bu for'u kullandık.
            for a in range(row+1): # İşlem yaptığı satırda kaç sayı yazdıracağımızı bu for ile belirledik.
                print(numbers,end="") # Sayılarımızı yazdırdık.
                print(" ",end="") # Aynı satırdaki sayılar arasında boşluk bıraktık.
                numbers+=1
                
            print("") #Alt satıra geçmek için kullandık.
```
