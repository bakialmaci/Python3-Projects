<b>Örnek 1 : IF-WHILE-FOR</b><br><br>

Bu örnekte artık aşama aşama açıklamayı bir kenera bırakıyorum. Sizden şu adımları uygulayacak bir program yazmanızı istiyorum.<br><br>

<b>1)</b> Kullanıcıdan <span style="color:green"><b>integer input</b></span> alacak.<br>
<b>2)</b> Aldığı <span style="color:green"><b>input</b></span>'u değerlendirecek:<br>
&emsp;<b>a)</b> Eğer <span style="color:blue"><b>0</b></span>'ise <span style="color:darkred"><b>"Quit"</b></span> yazdırıp programdan çıkacak. <br>
&emsp;<b>b)</b> Eğer <span style="color:blue"><b>1</b></span>' ise Program  <span style="color:darkred"><b>"Geçersiz INPUT"</b></span> yazdıracak ve <span style="color:red"><b>tekrar input almaya devam edecek.</b></span><br> 
&emsp;<b>c)</b> Eğer <span style="color:purple"><b>0'dan küçük</b></span> veya <span style="color:purple"><b>20'den büyük</b></span> bir <span style="color:green"><b>input</b></span> gelirse <span style="color:darkred"><b>"OVERFLOW"</b></span> yazdıracak ve <span style="color:red"><b>tekrar input almaya devam edecek.</b></span> <br>
&emsp;<b>d)</b> Diğer tüm durumlarda <span style="color:blue"><b>girdiği satır sayısında</b></span> <b>örnekteki üçgeni</b> ekrana yazdıracak ve <span style="color:red"><b>tekrar input almaya devam edecek.</b></span> <br>
&emsp;<b>ÜÇGEN ÖRNEĞİ : </b> input = 4 <br>
&emsp;&emsp; 1<br>
&emsp;&emsp; 2 3<br>
&emsp;&emsp; 4 5 6<br>
&emsp;&emsp; 7 8 9 10<br>






<b>3)</b> **<span style="color:darkred"><b>Programdan çıkma emri (input = 0) gelene kadar program sürekli çalışmaya devam edecek.!</b></span>**<br>

***Karakterleri yanyana yazmak için print fonksiyonunun her seferinde yeni satıra geçmesini bu şekilde çözebilirsiniz => <span style="color:green">print(a, end = '')</span>**<br>
***Python'da <span style="color:green">print("")</span> fonksiyonu alt satira geçmek için kullanılır.**

```python
while 1:
    
    inp = int(input("Satır sayısını giriniz: ")) #  Kullanıcıdan integer bir input aldık.
    
    if inp == 0:
        print("Quit")
        break
    elif inp == 1:
        print("Geçersiz INPUT")
    elif (inp < 0 or inp > 20):
        print("OVERFLOW")
    else:
        
        numbers = 1 # Üçgeni 1'den başlaktık.

        for row in range(inp): # Input kadar satır yazdırmak için bu for'u kullandık.
            for a in range(row+1): # İşlem yaptığı satırda kaç sayı yazdıracağımızı bu for ile belirledik.
                print(numbers,end="") # Sayılarımızı yazdırdık.
                print(" ",end="") # Aynı satırdaki sayılar arasında boşluk bıraktık.
                numbers+=1
                
            print("") #Alt satıra geçmek için kullandık.
```

İhtiyacım olan özellikler; Kullanıcıların isim,soyisim,doğum yılı ve hastalık nedeninin bulunduğu 4 adet liste oluşturman. Ayrıca girilen inputlar her ne olursa olsun isim ve soyisimlerin **sadece baş harfleri** her zaman büyük olmalıdır. (lower(), capitalize() ve title() methodlarını kullanabilirsiniz.)<br><br>

Bu soruda kopya vermiyorum. Yapmak istediğin işlemleri fonksiyonlara bölersen kolayca yapabilirsin. Ayrıca programın en dışında while koymayı unutma!.<br>

<b>Dikkat etmeniz gerekenler:</b><br>
<b>1)</b> Kullandığınız değişkenlerin fonksiyonların içinde değişip değişmediğini kontrol edin!<br>
<b>2)</b> Slaytlarda olmayan bir bilgi olduğunu düşünüyorsanız Hocaya sorun.<br>
<b>3)</b> hasta_arama fonksiyonunu sizin için hazırladım. İhtiyacınız olursa common_funcs modülünden fonksiyonu kodunuza ekleyebilirsiniz. (Kullanımı => check, index = hasta_arama(names) şeklindedir. check değişkeni True ise hasta kayıtlarda bulunuyor. index ise listenin kaçıncı elemanında o kişinin bulunduğunu söylüyor.<br>

```python
def hasta_kayit(names,surnames,birthdates,diagnosis):
    name = (input("Hasta İsmi: ").lower()).capitalize() # Alınan input u Baş harfleri büyük formata getirdik.
    surname = (input("Hasta Soyismi: ").lower()).capitalize() # Alınan input u Baş harfleri büyük formata getirdik.
    birth = input("Doğum Tarihi: ")
    diagnose = (input("Teşhis: ").lower()).title() # Alınan input u Baş harfleri büyük formata getirdik.
    
    names.append(name) # İsimler listesine hastanın ismini ekler.
    surnames.append(surname)
    birthdates.append(birth)
    diagnosis.append(diagnose)
    
def hasta_taburcu(names,surnames,birthdates,diagnosis):  
    check, index = hasta_arama(names)
    if check: # Hasta kayıtlı ise listeden silmek için bu bloğa girer.
        del names[index]
        del surnames[index]
        del birthdates[index]
        del diagnosis[index]
        print("Hasta Başarıyla Taburcu Edilmiştir.")  
    

def hasta_arama(names):
    name = (input("Hasta İsmi: ").lower()).capitalize() # Alınan input u Baş harfleri büyük formata getirdik.
    checkPatient = name in names # names listesinde name değişkeninin olup olmadığını kontrol edip boolean sonuç döndürür.
    
    if checkPatient: # Hasta var ise bu bloğa girer.
        index = names.index(name) # Hastanın bulunduğu index numarasını bulur.
        print("Hasta hastanemizde kayıtlıdır.")
    else:
        index = -1
        print("Hasta hastanemizde kayıtlı değildir.")
    
    return checkPatient,index

def hasta_guncelleme(names,surnames,birthdates,diagnosis):
    check, index = hasta_arama(names)
    
    if check:
        names[index] = (input("Hasta İsmi: ").lower()).capitalize() # Alınan input u Baş harfleri büyük formata getirdik.
        surnames[index] = (input("Hasta Soyismi: ").lower()).capitalize() # Alınan input u Baş harfleri büyük formata getirdik.
        birthdates[index] = input("Doğum Tarihi: ")
        diagnosis[index] = (input("Teşhis: ").lower()).title() # Alınan input u Baş harfleri büyük formata getirdik.
        print("Hasta başarıyla güncellendi.")
        
def hasta_goruntule(names,surnames,birthdates,diagnosis):
    if len(names) != 0: # Hasta isimlerinin boş olup olmadığını kontrol ettik. Boş ise ekrana Hata yazdırdık.
        for i in range(len(names)): # 0'dan başlayarak listedeki hasta sayısına kadar tüm hastaların indexlerini aldık.
            print(str(i+1) + ") " + names[i] + " - " + surnames[i] + " - " + birthdates[i] + " - " + diagnosis[i])
    else:
        print("Hasta bulunmamaktadır.")
        

def menu():
    print("----------------------------")
    print("1) HASTA KAYIT")
    print("----------------------------")
    print("2) HASTA TABURCU")    
    print("----------------------------")
    print("3) HASTA ARA")
    print("----------------------------")
    print("4) HASTA GÜNCELLE")
    print("----------------------------")
    print("5) HASTALARI GÖRÜNTÜLE")
    print("----------------------------")
    print("0) ÇIKIŞ")
    print("----------------------------")

def main():
    names = [] # Program ilk açıldığında Veri Tabanı olarak kullanılacak değişkenleri tanımladık.
    surnames = []
    birthdates = []
    diagnosis = []
    while 1: # Programdan çıkana kadar çalışmasını sağladık.
        menu()
        islem = int(input("-) İşlem Numarası :")) # İşlemi alıp integer'a çevirdik.

        if islem == 0:
            break
        elif islem == 1:
            hasta_kayit(names,surnames,birthdates,diagnosis)
        elif islem == 2:
            hasta_taburcu(names,surnames,birthdates,diagnosis)
        elif islem == 3:
            hasta_arama(names)
        elif islem == 4:
            hasta_guncelleme(names,surnames,birthdates,diagnosis)
        elif islem == 5:
            hasta_goruntule(names,surnames,birthdates,diagnosis)
        else:
            print("Hatalı İşlem Numarası")
            
main()
```
